```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two vectors with 15 random floats between 0 and 1
V1 = np.random.rand(15)
V2 = np.random.rand(15)
print("Vector 1:\n", V1)
print("\n Vector 2:\n", V2)

# Creating  index for the x-axis
x = np.arange(len(V1))
print("\n Index:",x)

# Plot the vectors
plt.plot(x, V1, marker='o', linestyle='-', label="Vector 1")
plt.plot(x, V2, marker='s', linestyle='--', label="Vector 2")

# Labels and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Two Random Vectors")
plt.legend()
plt.grid(True)


plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a 4x4 matrix with random values
matrix = np.random.rand(4, 4)  # Values between 0 and 1

# Define row and column labels
labels = ['A', 'B', 'C', 'D']

# Plot the heatmap
sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5, xticklabels=labels, yticklabels=labels)
plt.title('Heatmap of 4x4 Matrix')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two random 4x4 matrices with integer values between 0 and 10
A = np.random.randint(0, 10, size=(4, 4))  # Random integers between 0 and 9
B = np.random.randint(0, 10, size=(4, 4))  # Random integers between 0 and 9

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

# Bar plot to visualize results
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # Create figure and subplot array

# Define bar indices
indices = range(1, 17)  # 4x4 matrix flattened has 16 elements

# Sum
axes[0].bar(indices, matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(indices, matrix_diff.flatten(), color='black')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(indices, matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()  # Adjust layout for better readability
plt.show()
```
